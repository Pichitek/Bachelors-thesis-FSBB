```python
import csv
import numpy as np
import matplotlib.pyplot as plot
```

```python
############ input filter, not damped ############

filename = "input_flt_undamped.csv"

fields = []
rows = []

with open(filename, 'r') as csvfile:
    csvreader = csv.reader(csvfile,delimiter=";")
    fields = next(csvreader)
    for row in csvreader:
        rows.append(row)

# data from csv into vectors
VECTOR_SIZE = 7001

freq = np.zeros(VECTOR_SIZE)
mag = np.zeros(VECTOR_SIZE)
ang = np.zeros(VECTOR_SIZE)

i=0
for row in rows[:VECTOR_SIZE]:
    freq[i] = row[0]
    mag[i] = row[1]
    ang[i] = row[2]
    i+=1

i=0
for i in range(len(mag)):
    if np.round(mag[i]) == -3:
        break
    i+=1

plot.figure(figsize=(10,7))
plot.semilogx(freq, mag, label='Wzmocnienie filtru')
plot.plot(freq[i],-3,'ro', label='Częstotliwość odcięcia')
plot.legend()
plot.xlabel("Częstotliwość [Hz]")
plot.ylabel("Wzmocnienie [dB]")
plot.grid()
```

```python
############ single ended to differential ############
filename = "singleended_to_diff.csv"

fields = []
rows = []

with open(filename, 'r') as csvfile:
    csvreader = csv.reader(csvfile,delimiter=";")
    fields = next(csvreader)
    for row in csvreader:
        rows.append(row)

# data from csv into vectors
VECTOR_SIZE = 655

time_sym= np.zeros(VECTOR_SIZE)
Vin = np.zeros(VECTOR_SIZE)
n = np.zeros(VECTOR_SIZE)
n_p = np.zeros(VECTOR_SIZE)
p = np.zeros(VECTOR_SIZE)

i=0
for row in rows[:VECTOR_SIZE]:
    time_sym[i] = row[0]
    Vin[i] = row[1]
    n[i] = row[2]
    n_p[i] = row[3]
    p[i] = row[4]
    i+=1


plot.figure(figsize=(14,11))
axs[0,0] = plot.subplot(3,1,1)
axs[0,0].plot(time_sym, Vin, label='Sygnał wejściowy konwertera')
axs[0,0].legend()
axs[0,0].set_xlabel("Czas [s]")
axs[0,0].set_ylabel("Napięcie [V]")
axs[0,0].set_xlim([0,0.0001])
axs[0,0].set_ylim([1.5,3.5])
axs[0,0].grid()

axs[0,1] = plot.subplot(3,1,2)
axs[0,1].plot(time_sym, n, label='Sygnał negatywny, względem masy')
axs[0,1].plot(time_sym, p, label='Sygnał pozytywny, względem masy')
axs[0,1].legend()
axs[0,1].set_xlabel("Czas [s]")
axs[0,1].set_ylabel("Napięcie [V]")
axs[0,1].set_xlim([0,0.0001])
axs[0,1].set_ylim([0,5])
axs[0,1].grid()

axs[0,2] = plot.subplot(3,1,3)
axs[0,2].plot(time_sym, n_p, label='Sygnał różnicowy')
axs[0,2].legend()
axs[0,2].set_xlabel("Czas [s]")
axs[0,2].set_ylabel("Napięcie [V]")
axs[0,2].set_xlim([0,0.0001])
axs[0,2].set_ylim([1.5,3.5])
axs[0,2].grid()
```

```python
############ Buck efficiency graph ############

filename = "Buck_eff.csv"

fields = []
rows = []

with open(filename, 'r') as csvfile:
    csvreader = csv.reader(csvfile,delimiter=";")
    fields = next(csvreader)
    for row in csvreader:
        rows.append(row)

# data from csv into vectors
VECTOR_SIZE = 11

Bu_pin = np.zeros(VECTOR_SIZE)
Bu_pout = np.zeros(VECTOR_SIZE)
Bu_n = np.zeros(VECTOR_SIZE)

i=0
for row in rows[:VECTOR_SIZE]:
    Bu_pin[i] = row[0]
    Bu_pout[i] = row[1]
    Bu_n[i] = row[2]
    i+=1

plot.figure(figsize=(10,7))
plot.plot(Bu_pout, Bu_n, 'b-o')
plot.legend()
plot.xlabel("Moc wyjściowa [W]")
plot.ylabel("Sprawność[%]")
plot.ylim([0.8,1])
plot.xlim([0,36])
plot.grid()
```
